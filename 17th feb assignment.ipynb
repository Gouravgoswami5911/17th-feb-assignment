```python
#Question4: Using the database and the collection created in question number 3, write a code to insert one record,
# and insert many records. Use the find() and find_one() methods to print the inserted record.




# Import the necessary libraries
import pymongo

# Connect to MongoDB
client = pymongo.MongoClient('mongodb://localhost:27017/')

# Select the database and collection
mydb = client['mydatabase']
mycol = mydb['customers']

# Insert one record
mydict = { "name": "John", "address": "Highway 37" }
x = mycol.insert_one(mydict)
print("Inserted ID:", x.inserted_id)

# Insert many records
mylist = [
  { "name": "Amy", "address": "Apple st 652" },
  { "name": "Hannah", "address": "Mountain 21" },
  { "name": "Michael", "address": "Valley 345" },
  { "name": "Sandy", "address": "Ocean blvd 2" },
  { "name": "Betty", "address": "Green Grass 1" },
  { "name": "Richard", "address": "Sky st 331" },
  { "name": "Susan", "address": "One way 98" },
  { "name": "Vicky", "address": "Yellow Garden 2" },
  { "name": "Ben", "address": "Park Lane 38" },
  { "name": "William", "address": "Central st 954" },
  { "name": "Chuck", "address": "Main Road 989" },
  { "name": "Viola", "address": "Sideway 1633" }
]
x = mycol.insert_many(mylist)
print("Inserted IDs:", x.inserted_ids)

# Use find() and find_one() methods to print the inserted records
print("One record:")
print(mycol.find_one())
print("All records:")
for record in mycol.find():
    print(record)
```

```python
#Queston5: Explain how you can use the find() method to query the MongoDB database. Write a simple code to
# demonstrate this.



The find() method in MongoDB is used to query the database and retrieve documents that match a specified query criteria. It 
returns a cursor object that can be iterated to access all matching documents.

The find() method takes an optional parameter that specifies the query criteria. The query criteria can be specified using a 
document that contains one or more field-value pairs. The field represents the name of the field in the document and the value
represents the value of the field that we want to match. The find() method returns all documents that match the specified query
criteria.

Here is an example code that demonstrates how to use the find() method in MongoDB to retrieve documents that match a specific
query criteria:

# Import the necessary libraries
import pymongo

# Connect to MongoDB
client = pymongo.MongoClient('mongodb://localhost:27017/')

# Select the database and collection
mydb = client['mydatabase']
mycol = mydb['customers']

# Insert some sample data
mylist = [
  { "name": "John", "address": "Highway 37" },
  { "name": "Amy", "address": "Apple st 652" },
  { "name": "Hannah", "address": "Mountain 21" },
  { "name": "Michael", "address": "Valley 345" }
]
x = mycol.insert_many(mylist)

# Use the find() method to query the database
query = { "address": "Mountain 21" }
result = mycol.find(query)

# Print the matching documents
for doc in result:
    print(doc)
```

```python
 #Question6: Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
    
    
The sort() method in MongoDB is used to sort the results of a query in ascending or descending order based on one or more 
fields in the documents. The sort() method takes an optional parameter that specifies the sorting criteria. The sorting 
criteria can be specified using a document that contains one or more field-value pairs. The field represents the name of the 
field in the document, and the value represents the sorting order (either 1 for ascending order or -1 for descending order).

Here's an example code to demonstrate sorting in MongoDB:   



# Import the necessary libraries
import pymongo

# Connect to MongoDB
client = pymongo.MongoClient('mongodb://localhost:27017/')

# Select the database and collection
mydb = client['mydatabase']
mycol = mydb['customers']

# Insert some sample data
mylist = [
  { "name": "John", "address": "Highway 37", "age": 25 },
  { "name": "Amy", "address": "Apple st 652", "age": 30 },
  { "name": "Hannah", "address": "Mountain 21", "age": 20 },
  { "name": "Michael", "address": "Valley 345", "age": 35 }
]
x = mycol.insert_many(mylist)

# Use the find() method with the sort() method to query the database and sort the results
result = mycol.find().sort("age", -1)

# Print the sorted documents
for doc in result:
    print(doc)
```
